```python
import pandas as pd
import numpy as np

dataset=pd.read_csv('India.csv').values

data=dataset[:,0].reshape(-1,1)
target=dataset[:,1].reshape(-1,1)

from matplotlib import pyplot as plt

plt.plot(data,target)
plt.xlabel('Days')
plt.ylabel('We')
plt.title('wbD')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(degree=4,include_bias=False)
data_new=poly.fit_transform(data)
```

```python
from sklearn.linear_model import LinearRegression

algorithm=LinearRegression()
algorithm.fit(data_new,target)

print('Coefficients:',algorithm.coef_)
print('Intercept:',algorithm.intercept_)
```

```python
from sklearn.metrics import r2_score

predicted_target=algorithm.predict(data_new)

print('r2 score:',r2_score(target,predicted_target))
```

```python
plt.plot(data,target,'b',label='Actual')
plt.plot(data,predicted_target,'g--',label='Regression Curve')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Confirmed Patients')
plt.title('India 30-01 to 05-05')
plt.show()
```

```python
test_data=np.arange(94,124).reshape(-1,1)
test_data_new=poly.fit_transform(test_data)
```

```python
predicted_target_next_10_days=algorithm.predict(test_data_new)
```

```python
plt.plot(data,target,'b',label='Actual: From 30-01 to 05-05')
plt.plot(test_data,predicted_target_next_10_days,'r',label='Prediction: From 06-05 to 05-06')



plt.legend()
plt.xlabel('Days')
plt.ylabel('Deceased Patients')
plt.title('Gujarat')
plt.savefig("gujD")
plt.show()
```
